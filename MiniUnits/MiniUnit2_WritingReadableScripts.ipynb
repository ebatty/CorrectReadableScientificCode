<details>
<summary> <font color='blue'>Click here for solution explanation </font></summary>

There are a few things making this snippet less readable:
*  Unclear variable names
*  Having everything on one line
*  Not obeying Python style guide conventions for whitespaces around operators
* The else clause isn't necessary if we use >= or <=

We can fix these issues with:



```
a = 10
b = 4

# Computes the absolute difference
if a >= b:
  absolute_difference = a - b
elif a < b:
  absolute_difference = b - a
```

Even better though, we could use the absolute value function that already exists in Python:

```
a = 10
b = 4

# Computes the absolute difference
absolute_difference = abs(a - b)
```

````python
# @markdown Click here to see helper functions

def checks_for_numbers(password):
    return any(char.isdigit() for char in password)

def checks_for_letters(password):
  return any(char.isalpha() for char in password)
````

````python
# Update the code here

bb = 'ilikedogs' # prior password (will check it is not the same)
p = 'ilovedogsalot' # password to check

# Check if password is good to use
if type(p)==str:
    if len(p)>=12:
        if p!=bb:
            n = checks_for_numbers(p)
            if n==True and checks_for_letters(p):
               print('Password is great')
            else:
                print('Not usable: needs both letters and numbers')
        else:
            print('Not usable: password is same as previous one')
    else:
        print('Not usable: password is too short')
else:
    print('Not usable: password is not a string')
````

<details>
<summary> <font color='blue'>Click here for solution explanation </font></summary>

There are quite a few aspects that made this code less readable, including:
* Bad variable names, including bad Boolean names
    - We will rename `bb` to `prior_password`, `p` to `password`, `n` to `has_numbers`
* Use of magic number 12
  - We will create a constant that is the min password length
* Over nesting
   - We can fix using guard clauses, by checking conditions that make the password unusable and exiting out if one is hit
* Lack of consistency in when the password is checked for numbers vs letters
   - We will create another variable `has_letters`
* Comparing a boolean value to True (not necessary)
  - We can change ` if l==True` to `if l`
* Does not use Python style guide recommendation of white spaces around operators
* Very advanced: it is more Pythonic to use `isinstance(password, str)`, instead of `type(p) == str`.

This is one example fix:

```
MIN_PASSWORD_LENGTH = 12

prior_password = 'ilikedogs'
password = 'ilovedogsalot'

# Check if password is good to use
has_letters = checks_for_letters(password)
has_numbers = checks_for_numbers(password)

if not isinstance(p, str):
    print('Not usable: password is not a string')
elif len(p) < MIN_PASSWORD_LENGTH:
    print('Not usable: password is too short')
elif p == bb:
    print('Not usable: password is same as previous one')
elif not has_letters and not has_numbers:
    print('Not usable: needs both letters and numbers')
else:
    print('Password is great')
```
